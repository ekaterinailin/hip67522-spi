```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
flares = pd.read_csv('../results/flare_phases_and_energies.csv')
flares = flares[np.cos(flares.phase * 2 * np.pi) > 0]
```

```python
newed_rec = flares.ed_rec / np.cos(flares.phase * 2 * np.pi)
phaseoff = (flares.phase + 0.5)% 1 - 0.5
```

```python
plt.scatter(phaseoff, newed_rec, c='r')
plt.scatter(phaseoff, flares.ed_rec)
plt.yscale('log')
plt.xlim(-0.5, 0.5)
plt.xlabel('Orbital phase')
plt.ylabel('Flare energy (erg)')
```

```python
np.sum(newed_rec) / np.sum(flares.ed_rec)
```
